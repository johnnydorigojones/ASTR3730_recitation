# 5
Using the distance $D$ to the object (already defined above in units of pc), estimate the bolometric luminosity.

--Calculate $L_{\lambda}$ at the wavelength where the continuum peaks, $L_{\lambda_{\rm max}}$. You will have to estimate the flux density, $F_\lambda$, at $\lambda_{max}$ on the plot.

--Estimate the bolometric luminosity as $L_{bol} = \int d\lambda L_{\lambda} \approx \lambda_{\rm max} L_{\lambda_{\rm max}}$

```python
# convert the distance initially given above in pc to cm
D1 = D_obj1*3.086e+18 # 1.5 parsecs in cm
D2 = D_obj2*3.086e+18 # 5.1e4 pc, or 51 kpc, in cm
D3 = D_obj3*3.086e+18 # 68 pc in cm

# flux density at the wavelength where the continuum peaks in units of ergs/cm^2/s/Angstrom
# multiply by 10^-10 because we initially divided by 10^-10 to scale the y-axis
F_lambdamax_obj1 = 40*1e-10 
F_lambdamax_obj2 = 0.6*1e-10
F_lambdamax_obj3 = 0.035*1e-10

# convert F from ergs/cm^2/s/Angstrom to ergs/cm^3/s (i.e. cancel out the Angstrom unit)
# we can actually keep F in units ergs/cm^2/s/Angstrom, but only if lambda_max is in Angstrom when calculating L_bol
F_lambdamax_obj1 = F_lambdamax_obj1*1e8
F_lambdamax_obj2 = F_lambdamax_obj2*1e8
F_lambdamax_obj3 = F_lambdamax_obj3*1e8

# convert lambda_max from Angstrom to cm
lambda_max_cm_obj1 = lambda_max_angstrom_obj1/1e8
lambda_max_cm_obj2 = lambda_max_angstrom_obj2/1e8
lambda_max_cm_obj3 = lambda_max_angstrom_obj3/1e8

L_lambdamax_obj1 = 4*np.pi*(D1**2)*F_lambdamax_obj1 # ENTER EXPRESSION IN TERMS OF D AND F_LAMBDAMAX
Lbol_obj1 = lambda_max_cm_obj1*L_lambdamax_obj1 # ENTER APPROXIMATE EXPRESSION GIVEN ABOVE

L_lambdamax_obj2 = 4*np.pi*(D2**2)*F_lambdamax_obj2 # ENTER EXPRESSION IN TERMS OF D AND F_LAMBDAMAX
Lbol_obj2 = lambda_max_cm_obj2*L_lambdamax_obj2 # ENTER APPROXIMATE EXPRESSION GIVEN ABOVE

L_lambdamax_obj3 = 4*np.pi*(D3**2)*F_lambdamax_obj3 # ENTER EXPRESSION IN TERMS OF D AND F_LAMBDAMAX
Lbol_obj3 = lambda_max_cm_obj3*L_lambdamax_obj3 # ENTER APPROXIMATE EXPRESSION GIVEN ABOVE

print('Bolometric luminosity of object 1 in units L_Sun:', round(Lbol_obj1/3.846e33, 2))
print('Bolometric luminosity of object 2 in units L_Sun:', round(Lbol_obj2/3.846e33, 2))
print('Bolometric luminosity of object 3 in units L_Sun:', round(Lbol_obj3/3.846e33, 2))
print()
```

# 6
Using your results for $L$ and $T$, use $T = T_{\rm eff}$ to estimate the radius $R$ of the object

```python
Teff_obj1 = T_obj1
Teff_obj2 = T_obj2
Teff_obj3 = T_obj3

sigma_b = 5.6704e-5 # Stefan-Boltzmann constant in units g/s^3/K^4

R_obj1 = np.sqrt(Lbol_obj1/(4*np.pi*sigma_b*Teff_obj1**4)) # EXPRESSION IN TERMS OF LBOL AND TEFF
R_obj2 = np.sqrt(Lbol_obj2/(4*np.pi*sigma_b*Teff_obj2**4)) # EXPRESSION IN TERMS OF LBOL AND TEFF
R_obj3 = np.sqrt(Lbol_obj3/(4*np.pi*sigma_b*Teff_obj3**4)) # EXPRESSION IN TERMS OF LBOL AND TEFF

print('radius of object 1 in Solar Radii:', round(R_obj1/6.957e+10,3))
print('radius of object 2 in Solar Radii:', round(R_obj2/6.957e+10,3))
print('radius of object 3 in Solar Radii:', round(R_obj3/6.957e+10,3))
```
